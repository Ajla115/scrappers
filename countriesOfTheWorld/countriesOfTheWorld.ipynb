```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.scrapethissite.com/pages/simple/"

page = requests.get(url)
soup = BeautifulSoup(page.text, "html")

soup
```

```python
countryNames = soup.find_all("h3", class_ = "country-name")
countryNamesList = []
for name in countryNames:
    countryNamesList.append(name.text.strip())

countryNamesList


    
```

```python
countryInfo = soup.find_all("div", class_ = "country-info")
countryInfo
```

```python
capitalList = []

for itemsList in countryInfo:
    capitalList.append(itemsList.find("span", class_ = "country-capital").text.strip())

capitalList
```

```python
countryCapital = soup.find_all("span", class_ = "country-capital")
countryPopulation = soup.find_all("span", class_ = "country-population")
countryArea = soup.find_all("span", class_ = "country-area")
```

```python
populationList = []

for itemsList in countryInfo:
    populationList.append(itemsList.find("span", class_ = "country-population").text.strip())

populationList
```

```python
areaList = []

for itemsList in countryInfo:
    areaList.append(itemsList.find("span", class_ = "country-area").text.strip())

areaList
```

```python
d = {

    "Country Name" : [ element for element in countryNamesList],
    "Capital City" : [ element for element in capitalList],
    "Population" : [ element for element in populationList],
    "Area" : [ element for element in areaList ] 
}

df = pd.DataFrame(data=d)
df
```

```python
#performing analysis
#checking how many countries in their name have 1 word, 2 words, 3 words or more

df["Word Count"] = df["Country Name"].apply(lambda x: len(x.split()))

word_count_distribution = df["Word Count"].value_counts()

word_count_distribution
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
word_count_distribution.plot(kind='bar')
plt.xlabel("Number of Words in Country Name")
plt.ylabel("Count of Countries")
plt.title("Distribution of Countries by Name Length")
plt.show()
```

```python
df = df.drop('Word count', axis=1)  
df
```

```python
#calculate density of an area and add it as a new column

#df["Population"].astype(int).dtypes
#df["Area"].astype(float).dtypes
#df.astype({'Population': 'float'}).dtypes

df["Population"] = pd.to_numeric(df["Population"], errors="coerce").fillna(0).astype(int)
df["Area"] = pd.to_numeric(df["Area"], errors="coerce").fillna(1).astype(float)

#errors="coerce" will force conversion

df.info()
```

```python
df["Region Density"] = df["Population"] / df["Area"]
df
```

```python
#round the result for region density
df = df.round({"Region Density": 2})
df
```

```python
#check if there are duplicate values in any column
duplicate_count = df.duplicated().sum()  
duplicate_count
```

```python
#check the number of cities that end in every vowel (a, i, e, o, u)

countDict = {
    "a" : 0,
    "o" : 0,
    "e" : 0,
    "u" : 0,
    "i" : 0
}

for city in df["Capital City"]:
    if city[-1] == 'a':
        countDict["a"] += 1
    elif city[-1] == 'e':
        countDict["e"] += 1
    elif city[-1] == 'i':
        countDict["i"] += 1
    elif city[-1] == 'o':
        countDict["o"] += 1
    else:
        countDict["u"] += 1
        
keysList = []
valuesList = []

for key in countDict.keys():
    keysList.append(key)

for value in countDict.values():
    valuesList.append(value)

fig, ax = plt.subplots()
ax.pie(valuesList, labels=keysList, autopct='%1.1f%%');  
```
